```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from pandas.api.types import CategoricalDtype

plt.rcParams['figure.figsize'] = [12, 7]
sns.set()
```

```python
Zambia_file = Path('/storage/earth_data/home/ee21xs/XiaoShu/Dissertation/Data_Zambia.xlsx')
df_Zambia = pd.read_excel(Zambia_file, index_col=0, engine='openpyxl')
display(df_Zambia.head())
```

# Gender & variables

## Age

```python
contingency_table_age=pd.crosstab(df_Zambia["Age group"],df_Zambia["Gender"])
display(contingency_table_age)
```

```python
chi2_age, p_age, dof_age, expected_age = stats.chi2_contingency(contingency_table_age)
print('Chi Square Statistic = '+str(chi2_age))
print('p-value = '+str(p_age))
print('Degrees of Freedom: '+str(dof_age))
print('Expected frequencies: ')
display(pd.DataFrame(expected_age))
```
